## Problem 1

In this problem we will look at the space of real polynomials of degree $\leq n$:

$$
\mathbb{P}_n(\mathbb{R}) = \left\{f: \mathbb{R}\to\mathbb{R} \quad\biggr\rvert\quad f(x) = \sum_{k=0}^n a_k x^k \quad \textrm{for some }a_1, ..., a_n \in \mathbb{R}\right\}.
$$
Note that each point in the space is a function, namely a polynomial.
### a)
Write down the definition of a real vector space.

<font color ='blue'>
    Solution:
    
</font>

_Definition:_
A real vector space _V_ is a set with a multiplicative and an additive operator, which both are 1. commutative, 2. associative, 3. invertible, 4. have respective identity elements and 5. are distributible. In effect, for 

addition $(+)$: $\quad V \times V \rightarrow V: (u,v) \mapsto u+v,  \quad u,v\in V$,
1. $\quad u + v = v + u \quad \forall u,v \in V$
2. $\quad(u + v) + w = u + (v + w), \qquad \forall u,v,w \in V$
3. $\quad \exists w \in V\quad s.t. \quad v + w = 0 \quad \forall v \in V$
4. $\quad \exists 0\in V \quad s.t.\quad 0 + v = v \quad \forall v\in V$

multiplication $(\cdot)$: $\quad \mathbb{R} \times V \rightarrow V, \quad (\lambda,v)\mapsto \lambda \cdot v$,
1. $\quad (\lambda\mu)\cdot v = (\mu\lambda)\cdot v, \quad \forall \lambda,\mu \in \mathbb{R}, v \in V$
2. $\quad \lambda(\mu \cdot v) = (\lambda\mu)\cdot v \quad \forall \lambda,\mu \in \mathbb{R}, v \in V^*$
3. $\quad (\lambda \mu)\cdot v = (1)\cdot v \quad \forall v \in V^*$
4. $\quad 1 \cdot v = v \qquad \forall v \in V^*$
5. $\quad \lambda \cdot(u + v) = \lambda\cdot u + \lambda\cdot v \quad \forall u,v \in V, \lambda,\mu \in \mathbb{R}$

*_These properties follows to some degree from the properties of $\mathbb{R}$ under addition and multiplication._ 

### b)

Show that $\mathbb{P}_n(\mathbb{R})$ is a real vector space.

*Hint: Recall the how we define adding two functions, and how we define multiplying with a scalar.*

<font color ='blue'>
    Solution:
    
</font>

$x\in\mathbb{P}_n(\mathbb{R}) \Rightarrow \quad x = a_0 + a_1 x + a_2 x^2 + ... + a_{n-1} x^{n-1} + a_{n}x^n, \quad a_i,x\in\mathbb{R}, \forall i=0,1,... $
Another way to express the element $x\in\mathbb{P}_n(\mathbb{R})$ is like a sequence of coefficients $(a_i)_{i},$ with $a \in \mathbb{R}^{n+1} \quad \forall i=0,1,...,n-1,n$.

For any $x,y,z \in \mathbb{P}_n(\mathbb{R})\quad s.t. \quad x=(a_i)_{i}, \quad y=(b_i)_{i},\quad z=(c_i)_{i}, \quad i=0,1,...,n-1,n;\quad$ for any $\lambda,\mu \in \mathbb{R}:$

1. Commutativity: 
$$x + y = (a_i)_{i\in n} + (b_i)_{i\in n} = (a_i + b_i)_{i\in n} = (b_i+ a_i)_{i\in n} = y + x$$
$$ (\lambda\mu)\cdot x = (\lambda\mu)\cdot (a_i)_{i\in n} = (\lambda\mu a_i)_{i\in n} = (\mu\lambda a_i)_{i \in n} = (\mu\lambda)\cdot x$$
2. Associativity: 
$$(x + y) + z = \left((a_i)_{i} + (b_i)_{i}\right) + (c_i)_{i} = ((a_i + b_i) + c_i)_{i} = (a_i + (b_i + c_i))_{i} = x + (y + z) $$

3. Invertibility:
$$ -x = (-a_i)_i, \quad a_i \in \mathbb{R} \Rightarrow (-a_i)\in\mathbb{R} \quad \Rightarrow x + (-x) = (a_i)_i + (-a_i)_i = (a_i+(-a_i))_i = (0)_i = 0$$

4. Identity element:
$$0\in \mathbb{P}_n(\mathbb{R}) \quad s.t.\quad 0 = (0)_i, \quad i=0,1,2,...,n-1, n$$
$$\Rightarrow 0 + x = (0)_i + (a_i)_i = (0 + a_i)_i = (a_i)_i = x$$

$$1\in\mathbb{R}\quad \Rightarrow 1\cdot x = 1\cdot (a_i)_i = (1a_i)_i = (a_i)_i = x$$

5. Distributivity:
$$ \lambda \cdot (x + y) = \lambda \cdot (a_i + b_i)_i = (\lambda(a_i + b_i))_i = (\lambda a_i + \lambda b_i)_i = (\lambda a_i)_i + (\lambda b_i)_i = \lambda\cdot (a_i)_i + \lambda \cdot (b_i)_i = \lambda \cdot x + \lambda \cdot  y $$

$\Rightarrow \mathbb{P}_n(\mathbb{R})$ a vector space. $_\Box$

## c)

Show that $\mathbb{P}_n(\mathbb{R})$ has dimension $n+1$.

*Hint 1: Find a set of polynomials you think is a basis for $\mathbb{P}_n(\mathbb{R})$. Verify that it is in fact a basis.*

*Hint 2: If $p \in \mathbb{P}_n(\mathbb{R})$ and $p \neq 0$, how many roots can $p$ have at most? Recall the fundamental theorem of algebra.*


<font color ='blue'>
    Solution:
    
</font>

Let $S$ be a set of polynomials defined as $S = \left\{s_i \in \mathbb{P}_n(\mathbb{R}) | s_i = 1\cdot x^i, 0\leq i \leq n\right\}.$

In other words, the polynomial coefficients of $s_i$ are defined as $s_{ik} = \left\{\begin{array}{} 1, i=k \\ 0 i\neq k \end{array}\right..$

Any two elements $s_i,s_j\in S, i\neq j$ are linearly independent (or orthogonal), as the product of their coefficient vectors is $(s_i)^T s_j = 0$. 

Also, for any $q\in \mathbb{P}_n(\mathbb{R}): \qquad q =  \sum_{k=0}^n a_k x^k =  \sum_{k=0}^n a_k s_k \qquad \Rightarrow S$ linear basis for $\mathbb{P}_n(\mathbb{R}).$ 

As the coefficient vector consists of $n+1$ orthogonal elements, $\mathbb{P}_n(\mathbb{R})$ must have the dimension $n+1._\square$

### d)
Write down the definition of a norm on a vector space.

<font color ='blue'>
    Solution:
    
</font>

A norm $\|.\|$ on a vector space $V$ over $\mathbb{F}\in(\mathbb{R},\mathbb{C})$ is a function $\|.\|: V \rightarrow \mathbb{R}$ that is 1. positive, 2. homogenous and 3. satisfies the triangle inequality.

In other words, for any and all $u,v \in V, \quad \alpha \in \mathbb{F}:$
1. $$\|u\|\geq 0, \qquad \|u\| = 0 \Leftrightarrow u = 0$$
2. $$\|\alpha u \| = |\alpha|\cdot\|u\|$$
3. $$\|u+v\|\leq \|u\|+\|v\|$$

### e)
Let $x_0,...,x_n$ be $n+1$ different real numbers. For a polynomial $q \in \mathbb{P}_n(\mathbb{R})$ we now define
$$
\|q\|_P := \sum_{k=0}^n|q(x_k)|.
$$
Show that $\|\cdot\|_P$ is a norm on $\mathbb{P}_n(\mathbb{R}).$



<font color ='blue'>
    Solution:
    
</font>

For any $q,p \in \mathbb{P}_n(\mathbb{R}):$
1. Positivity
$$ |q(x_k)|\geq 0 \quad\forall x_k \qquad \Rightarrow \|q\|_P = \sum_{k=0}^n|q(x_k)| \geq 0,$$
$$ \|q\|_P = \sum_{k=0}^n|q(x_k)| = 0 \qquad \Leftrightarrow |q(x_k)|=0 \quad\forall x_k\in \mathbb{R} \qquad \Leftrightarrow q = 0.$$
2. Homogeneity
$$ \|\alpha q\|_P \quad= \sum_{k=0}^n|\alpha q(x_k)| \quad= \sum_{k=0}^n|\alpha|\cdot |q(x_k)|\quad = |\alpha|\sum_{k=0}^n|q(x_k)|\quad= |\alpha|\cdot \|q\|_P.$$
3. Triangle inequality

$$ \|p + q\|_P = \sum_{k=0}^n|p(x_k) + q(x_k)| \quad\leq\quad \sum_{k=0}^n\left(|p(x_k)|+|q(x_k)|\right) = \left(\sum_{k=0}^n|p(x_k)|\right) + \left(\sum_{k=0}^n|q(x_k)|\right) = \|p\|_P + \|q\|_P._\square$$

### f)
We now define 
$$
\|q\|_{C^n} := \max_{0\leq j \leq n}|q^{(j)}(0)|
$$
where $$q^{(j)} = \dfrac{d^jq}{dx^j}.$$

Show that $\|\cdot\|_{C^n}$ is also a norm on $\mathbb{P}_n(\mathbb{R})$.

<font color ='blue'>
    Solution:
    
</font>

For any $q\in\mathbb{P}_n(\mathbb{R}): \qquad q^{(j)}(0)= \left(\dfrac{d^jq}{dx^j}\right)(0) = j!\cdot a_j,\quad$ as all other elements are zeroed out by multiplying with zero or differentiation.

1. Positivity
$$ |q^{(j)}(0)|\geq 0 \quad \Rightarrow \|q\|_{C^n} = \max_{0\leq j \leq n}|q^{(j)}(0)| \geq 0,$$
$$ \|q\|_{C^n} = \max_{0\leq j \leq n}|q^{(j)}(0)| = 0 \qquad \Leftrightarrow \qquad q^{(j)}(0) = 0 \quad \forall j \qquad \Leftrightarrow \qquad a_j = 0 \quad \forall j \qquad \Leftrightarrow \qquad q = 0.$$
2. Homogeneity
$$ \|\alpha q\|_{C^n} = \max_{0\leq j \leq n}|\alpha q^{(j)}(0)| = \max_{0\leq j \leq n}\left(|\alpha|\cdot|q^{(j)}(0)|\right) = |\alpha| \max_{0\leq j \leq n}|q^{(j)}(0)| = |\alpha| \cdot \|q\|_{C^n}.$$
3. Triangle inequality
$$ \|p+q\|_{C^n}=  \max_{0\leq j \leq n}|p^{(j)}(0)+ q^{(j)}(0)| \quad \leq \quad \max_{0\leq j \leq n}\left(|p^{(j)}(0)|+ |p^{(j)}(0)|\right) \quad \leq \quad \max_{0\leq j \leq n}|p^{(j)}(0)| + \max_{0\leq j \leq n}|q^{(j)}(0)| = \|p\|_{C^n} + \|q\|_{C^n}._\square$$
